```python
import pandas as pd
```

```python
import seaborn as sns

data = pd.read_csv("study.csv")
data = data.drop(columns="Unnamed: 0")

data["att_cat"] = data['attention'].apply(lambda x: str(x) + ' ')

sns.displot(
    data, 
    x="ror_volatility", 
    hue="attention", 
    element='step'
)
```

```python
data.groupby('attention').mean().plot()
```

```python
data.groupby('attention').mean()['ror_volatility'].plot()
```
